```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Prol= pd.read_csv('Prosper loan.csv')
```

```python
Prol.drop(columns=['ListingCreationDate', 'CreditGrade', 'ListingKey', 'ClosedDate', 'BorrowerState','GroupKey', 'DateCreditPulled','FirstRecordedCreditLine', 'InquiriesLast6Months','TotalInquiries',	'CurrentDelinquencies','AmountDelinquent',	'DelinquenciesLast7Years',	'PublicRecordsLast10Years',	'PublicRecordsLast12Months',	'RevolvingCreditBalance',	'BankcardUtilization',	'AvailableBankcardCredit',	'TotalTrades', 'TradesNeverDelinquent (percentage)', 'TradesOpenedLast6Months','IncomeVerifiable',	'TotalProsperLoans', 'TotalProsperPaymentsBilled',	'OnTimeProsperPayments',	'ProsperPaymentsLessThanOneMonthLate',	'ProsperPaymentsOneMonthPlusLate',	'ProsperPrincipalBorrowed',	'ProsperPrincipalOutstanding',	'ScorexChangeAtTimeOfListing',	'LoanCurrentDaysDelinquent',	'LoanFirstDefaultedCycleNumber',	'LoanMonthsSinceOrigination',	'LoanNumber',	'LoanOriginationQuarter',	'MonthlyLoanPayment',	'LP_CustomerPayments',	'LP_CustomerPrincipalPayments',	'LP_InterestandFees',	'LP_ServiceFees',	'LP_CollectionFees',	'LP_GrossPrincipalLoss',	'LP_NetPrincipalLoss',	'LP_NonPrincipalRecoverypayments',	'PercentFunded',	'Recommendations',	'InvestmentFromFriendsCount',	'InvestmentFromFriendsAmount',	'Investors'],inplace=True)
```

```python
Prol.info()
```

```python
numeric_cols = ['BorrowerAPR', 'EstimatedEffectiveYield', 'EstimatedLoss', 'EstimatedReturn', 'ProsperRating (numeric)', 'ProsperScore', 'EmploymentStatusDuration', 'CreditScoreRangeLower','CreditScoreRangeUpper', 'CurrentCreditLines','OpenCreditLines','TotalCreditLinespast7years','DebtToIncomeRatio'] 
for col in numeric_cols:
    Prol[col].fillna(Prol[col].mean(), inplace=True)
categorical_cols = ['ProsperRating (Alpha)', 'Occupation', 'EmploymentStatus']
for col in categorical_cols:
    mode_val = Prol[col].mode()[0] 
    Prol[col].fillna(mode_val, inplace=True)
```

```python
Prol.info()
```

```python
Prol
```

```python
loan_status_counts = Prol['LoanStatus'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
loan_status_counts.plot(kind='bar', color='#9b2226')
plt.title('Distribution of LoanStatus')
plt.xlabel('Loan Status')
plt.ylabel('Number of Loans')
plt.xticks(rotation=45)
plt.show()

# Print the result
print(loan_status_counts)
```

```python
# Calculate the average BorrowerAPR
average_borrower_apr = Prol['BorrowerAPR'].mean()

# Plotting
plt.figure(figsize=(6, 4))
plt.bar('Average Borrower APR', average_borrower_apr, color='#9b2226')
plt.title('Average Borrower APR')
plt.ylabel('APR')
plt.ylim(0, 1)  # Assuming APR is between 0 and 1 (you can adjust based on your data range)
plt.show()

# Print the result
print(f'Average Borrower APR: {average_borrower_apr:.4f}')
```

```python
# Plotting the distribution of CreditScoreRangeLower
plt.figure(figsize=(10, 6))
plt.hist(Prol['CreditScoreRangeLower'], bins=20, color='#9b2226', edgecolor='black')
plt.title('Distribution of CreditScoreRangeLower')
plt.xlabel('Credit Score Range Lower Bound')
plt.ylabel('Number of Borrowers')
plt.show()

# Print the result
print(Prol['CreditScoreRangeLower'].describe())
```

```python
# Plotting the distribution of DebtToIncomeRatio
plt.figure(figsize=(10, 6))
plt.hist(Prol['DebtToIncomeRatio'], bins=30, color='#9b2226', edgecolor='black')
plt.title('Distribution of DebtToIncomeRatio')
plt.xlabel('Debt to Income Ratio')
plt.ylabel('Number of Borrowers')
plt.grid(True)
plt.show()

# Print the result
print(Prol['DebtToIncomeRatio'].describe())
```

```python
# Plotting the distribution of EmploymentStatus
employment_status_counts = Prol['EmploymentStatus'].value_counts()

plt.figure(figsize=(10, 6))
employment_status_counts.plot(kind='bar', color='#9b2226')
plt.title('Distribution of EmploymentStatus')
plt.xlabel('Employment Status')
plt.ylabel('Number of Borrowers')
plt.xticks(rotation=45)
plt.show()

# Print the result
print(employment_status_counts)
```

```python
# Plotting BorrowerAPR vs. LoanOriginalAmount
plt.figure(figsize=(10, 6))
plt.scatter(Prol['LoanOriginalAmount'], Prol['BorrowerAPR'], alpha=0.5, color='#9b2226')
plt.title('BorrowerAPR vs. LoanOriginalAmount')
plt.xlabel('Loan Original Amount')
plt.ylabel('Borrower APR')
plt.grid(True)
plt.show()

# Print the correlation between BorrowerAPR and LoanOriginalAmount
correlation = Prol['BorrowerAPR'].corr(Prol['LoanOriginalAmount'])
print(f'Correlation between BorrowerAPR and LoanOriginalAmount: {correlation:.4f}')
```

```python
# Plotting CreditScoreRangeLower vs. LoanStatus
plt.figure(figsize=(12, 6))
Prol.boxplot(column='CreditScoreRangeLower', by='LoanStatus', grid=False, notch=True, patch_artist=True, boxprops=dict(facecolor='#9b2226'))

plt.title('CreditScoreRangeLower vs. LoanStatus')
plt.suptitle('')  # Suppress the default title
plt.xlabel('Loan Status')
plt.ylabel('Credit Score Range Lower Bound')
plt.xticks(rotation=45)
plt.show()

# Print the result
print(Prol.groupby('LoanStatus')['CreditScoreRangeLower'].describe())
```

```python
# Filter out extremely high incomes for better visualization (optional)
Prol_filtered = Prol[Prol['StatedMonthlyIncome'] < 20000]

# Plotting EmploymentStatus vs. StatedMonthlyIncome
plt.figure(figsize=(12, 6))
Prol_filtered.boxplot(column='StatedMonthlyIncome', by='EmploymentStatus', grid=False, notch=True, patch_artist=True, boxprops=dict(facecolor='#9b2226'))

plt.title('StatedMonthlyIncome vs. EmploymentStatus')
plt.suptitle('')  # Suppress the default title
plt.xlabel('Employment Status')
plt.ylabel('Stated Monthly Income')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting DebtToIncomeRatio vs. LoanStatus
plt.figure(figsize=(12, 6))
Prol.boxplot(column='DebtToIncomeRatio', by='LoanStatus', grid=False, notch=True, patch_artist=True, boxprops=dict(facecolor='#9b2226'))

plt.title('DebtToIncomeRatio vs. LoanStatus')
plt.suptitle('')  # Suppress the default title
plt.xlabel('Loan Status')
plt.ylabel('Debt To Income Ratio')
plt.xticks(rotation=45)
plt.show()

# Print the result
print(Prol.groupby('LoanStatus')['DebtToIncomeRatio'].describe())
```

```python
# Plotting StatedMonthlyIncome vs. LoanOriginalAmount
plt.figure(figsize=(10, 6))
plt.scatter(Prol['StatedMonthlyIncome'], Prol['LoanOriginalAmount'], alpha=0.5, color='#9b2226')
plt.title('StatedMonthlyIncome vs. LoanOriginalAmount')
plt.xlabel('Stated Monthly Income')
plt.ylabel('Loan Original Amount')
plt.grid(True)
plt.show()

# Print the correlation result
correlation = Prol['StatedMonthlyIncome'].corr(Prol['LoanOriginalAmount'])
print(f'Correlation between StatedMonthlyIncome and LoanOriginalAmount: {correlation:.4f}')
```
